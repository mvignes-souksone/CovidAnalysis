```python
def filter_response(r):
    d = r['name']['common']
    
    if 'subregion' in r:
        e = r['subregion']
        my_dict = {"Country" : d, "Subregion" : e}
    else:
        my_dict = {"Country" : d, "Subregion" : 'NA'}
        
    return my_dict
    
```

```python
def filter_response(r):
    return {"Country": r['name']['common'], "Subregion" : r.get('subregion', 'NA')}
```

```python
li = []
for x in l3 :
    li.append(filter_response(x))
    
    
```
